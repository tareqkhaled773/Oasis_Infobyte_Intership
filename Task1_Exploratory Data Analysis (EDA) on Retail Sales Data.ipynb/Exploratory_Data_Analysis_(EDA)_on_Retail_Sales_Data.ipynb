##**Step 1: Import Required Libraries**

```python
#import important modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##**Step 2: Load the Dataset**

```python
df = pd.read_csv('/content/retail_sales_dataset.csv')
```

##**Step 3: Basic Data Cleaning**


```python
df.info()  # Show dataset structure
df["Date"] = pd.to_datetime(df["Date"])  # Convert Date column to datetime forma
```

##**Step 4: Verify Total Amount Calculation**


```python
# Step 4: Verify Total Amount Calculation
df["Computed_Total"] = df["Quantity"] * df["Price per Unit"]
if (df["Total Amount"] != df["Computed_Total"]).sum() == 0:
    print("✅ Total Amount matches Quantity * Price per Unit")
df.drop(columns=["Computed_Total"], inplace=True)  # Remove temp column
```

##**Step 5: Check for Duplicates**

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

##**Step 6: Descriptive Statistics**

```python
print("\nDescriptive Statistics:")
print(df[["Age", "Quantity", "Price per Unit", "Total Amount"]].describe())
```

##**Step 7: Analyze Sales Trends Over Time**

```python
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Total Amount"].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(marker="o", linestyle="-", color="b")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

##**Step 8: Analyze Customer Age & Gender Influence on Purchasing Behavior**

```python
plt.figure(figsize=(12,5))
sns.lineplot(x=df.groupby("Age")["Total Amount"].mean().index,
             y=df.groupby("Age")["Total Amount"].mean().values,
             marker="o", color="r")
plt.title("Average Spending by Age")
plt.xlabel("Age")
plt.ylabel("Avg Total Amount ($)")
plt.grid()
plt.show()
```

```python
df_gender_spending = df.groupby("Gender")["Total Amount"].sum()
plt.figure(figsize=(6,4))
sns.barplot(x=df_gender_spending.index, y=df_gender_spending.values, palette="coolwarm")
plt.title("Total Spending by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales ($)")
plt.show()
```

##**Step 9: Analyze Popular Product Categories**

```python
df_category_sales = df.groupby("Product Category")["Total Amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=df_category_sales.index, y=df_category_sales.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.show()
```

##**Step 10: Analyze Relationship Between Age, Spending, and Product Preferences**

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Product Category", y="Age", data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Age Distribution Across Product Categories")
plt.show()
```

##**Step 11: Check Seasonal Shopping Trends**

```python
df["Month"] = df["Date"].dt.month
seasonal_trends = df.groupby("Month")["Total Amount"].sum()

plt.figure(figsize=(10,5))
seasonal_trends.plot(marker="o", linestyle="-", color="g")
plt.title("Seasonal Trends in Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.grid()
plt.show()
```

##**Step 12: Analyze Purchasing Behavior Based on Quantity Bought**

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Quantity", y="Total Amount", data=df)
plt.title("Total Spending vs. Number of Items Purchased")
plt.xlabel("Number of Items")
plt.ylabel("Total Spending ($)")
plt.show()
```

##**Step 13: Analyze Price Distribution Within Each Product Category**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Product Category", y="Price per Unit", data=df, palette="Set2")
plt.xticks(rotation=45)
plt.title("Price Distribution Across Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Price per Unit ($)")
plt.show()
```
